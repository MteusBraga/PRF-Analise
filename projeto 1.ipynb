Carregando DataFrames
<h3>1)Tarefa</h3>
<p>É possivel carregar um arquivo compactado usando o parametro compression no read do pandas e colocando o tipo do arquivo</p>

```python
acidentes2018 = pd.read_csv('https://arquivos.prf.gov.br/arquivos/index.php/s/MaC6cieXSFACNWT/download',
                            encoding='latin1',
                            sep=';',
                            low_memory=False, compression='zip')
acidentes2018.shape 
```

```python
acidentes2019 = pd.read_csv('https://arquivos.prf.gov.br/arquivos/index.php/s/kRBUylqz6DyQznN/download',
                            encoding='latin1',
                            sep=';',
                            low_memory=False, compression='zip')
acidentes2019.shape
```

```python
acidentes2020 = pd.read_csv('https://arquivos.prf.gov.br/arquivos/index.php/s/jdDLrQIf33xXSCe/download',
                            encoding='latin1',
                            sep=';',
                            low_memory=False, compression='zip')
acidentes2020.shape
```

Concatenando os DataFrames

```python
#visto que as colunas sao iguais em ambos dataframes vou usar o concat para colocar um em baixo do outro
todos_acidentes=pd.concat([acidentes2018,acidentes2019, acidentes2020], ignore_index=True)
todos_acidentes
```

Explore os conjuntos de dados, respondendo as questões:<br>

<h3>2 Tarefa</h3>
<h4>a)</h4>Quantidades de acidentes em 2018, 2019 e 2020 respectivamente: 69295, 67446 e 63548.
<p><h4>b)</h4>Se cada coluna é uma variavel entao armazena 30 variaveis.</p>


```python
acidentes2018=acidentes2018.assign(acidentes=1)
acidentes2019=acidentes2019.assign(acidentes=1)
acidentes2020=acidentes2020.assign(acidentes=1)

qtd_acid18=acidentes2018['acidentes'].sum()
qtd_acid19=acidentes2019['acidentes'].sum()
qtd_acid20=acidentes2020['acidentes'].sum()
print(qtd_acid18)
print(qtd_acid19)
print(qtd_acid20)
```

```python
#criando uma nova coluna com valor padrao 1 para somar
todos_acidentes=todos_acidentes.assign(acidentes=1)
```

(c) quais as cinco cidades brasileiras onde mais ocorreram acidentes em rodovias federais?

<h3>2)</h3>
<h4>c)</h4> Cidades que correram mais acidentes:

```python

top5cidades=todos_acidentes[['municipio','acidentes']].groupby('municipio').sum()

top5cidades=top5cidades.sort_values(['acidentes'], ascending=False)

top5cidades.head(5)
```

3ª Tarefa<br>
Para cada indicação abaixo, construa um novo dataframe, salvando-o com a terminação .csv.<br>
(a) ranking de acidentes por estado para todos os anos (um só arquivo combinando a<br>
informação dos três anos);


<h3>3)Tarefa</h3>
<p><h4>a)</h4>criando e salvando dataframe com o ranking por Estado</p>

```python
rankingUF=todos_acidentes[['uf','acidentes']].groupby('uf').sum()
rankingUF=rankingUF.sort_values(['acidentes'], ascending=False)
rankingUF
```

```python
rankingUF.to_csv('rankingUF.csv',  header=True)
```

(b) acidentes por dia da semana para todos os anos (um só arquivo combinando a informação dos três anos)

<h3>3)</h3><h4>b)</h4>Criando o dataframe e salvando-o em .csv


```python
rankingPorDia=todos_acidentes[['dia_semana','acidentes']].groupby('dia_semana').sum()
rankingPorDia=rankingPorDia.sort_values(['acidentes'], ascending=False)
rankingPorDia
```

```python
#salvando
rankingPorDia.to_csv('rankingUF.csv',  header=True)
```
